```
def Lower(corpus):
  for i in range(len(corpus)):
    corpus[i] = corpus[i].lower() # lower all the letters
    # Remove whitespaces(\t,\n)
    corpus[i] = re.sub(r'\W', ' ', corpus[i])
    corpus[i] = re.sub(r'\s+', ' ', corpus[i]) # Remove whitespaces(\t,\n)
    return corpus
```

```
# Step 6a
from nltk.stem import WordNetLemmatizer
```

```
# Step 6b
import nltk
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```
# Step 6c Data Cleaning
def clean_lem(corpus):
  clean_tokens = []
  for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    #print(nltk.pos_tag(tokens))
    lemmatizer = WordNetLemmatizer()
    lem_sentence = []
    for token, tag in pos_tag(tokens):
      #token = re.sub(r'https?:\/\/\S+','', token)
      token = re.sub("(@[A-Za-z0-9_]+)","",token)
      if tag.startswith('NN'):
        pos = 'n'
      elif tag.startswith('VB'):
        pos = 'v'
      else:
        pos = 'a'
      lem_sentence.append(lemmatizer.lemmatize(token,pos))

      if len(token)>3 and token not in string.punctuation and token.lower() not in stop_words:
        clean_tokens.append(token.lower())
    #print(clean_tokens)
    print(lem_sentence)
  return clean_tokens
```

```
clean_tokens = clean_lem(corpus1)
```

```
# Create a dictionary of tokens (sentence is tokenized into words)
wordfreq1 = {}
for sentence in corpus1:
  tokens = nltk.word_tokenize(sentence)
  for token in tokens:
    if token not in wordfreq1.keys():
      wordfreq1[token] = 1
    else:
      wordfreq1[token] += 1
```

```
# Step 7: Create a dictionary of tokens (sentence is tokenized into words)
wordfreq = {}
for token in clean_tokens:
    if token not in wordfreq.keys():
      wordfreq[token] = 1
    else:
      wordfreq[token] += 1
```

```
import heapq
```

```
# Step 8: Top 190 most frequently occured tokens (in sorted order)
# Words in Bag of Model
mostfreq = heapq.nlargest(190, wordfreq, key = wordfreq.get)
mostfreq
```

```
# (Optional) Frequency of each token
from nltk import FreqDist
freq_dist = FreqDist(clean_tokens)
print(freq_dist.most_common(20))
```
